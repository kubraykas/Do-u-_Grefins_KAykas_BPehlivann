```python
class CBAMCalculator:
    def __init__(self, ets_price):
        self.ets_price = ets_price

    def normalize_code(self, code):
        return code.strip()

    def get_data_by_cn(self, cn_code):
        cn_code = self.normalize_code(cn_code)
        data = CN_CODE_DATABASE.get(cn_code)

        if data:
            return {
                "description": data["description"],
                "category": data["category"],
                "direct_ei": data["direct"],
                "indirect_ei": data["indirect"],
                "total_ei": data["total"]
            }
        else:
            return None

    def calculate(self, quantity, direct_ei, indirect_ei, foreign_carbon_price=0):
        total_ei = direct_ei + indirect_ei
        total_emission = quantity * total_ei
        certificates = total_emission

        cost = total_emission * self.ets_price
        adjusted_cost = cost - (total_emission * foreign_carbon_price)

        return {
            "total_ei": total_ei,
            "total_emission": total_emission,
            "certificates": certificates,
            "cbam_cost": cost,
            "cbam_cost_adjusted": adjusted_cost
        }
```

```python
"""
‚úÖ PROFESSIONAL CBAM EXECUTIVE REPORT GENERATOR - GEMINI POWERED

Bu mod√ºl, t√ºm CBAM analizlerini birle≈ütirerek Gemini ile
profesyonel bir y√∂netici raporu olu≈üturur.
"""

import pandas as pd
from datetime import datetime

# ------------------------------
# 1Ô∏è‚É£ LLM CBAM √áƒ±ktƒ±sƒ±nƒ± Parse Et
# ------------------------------
def parse_llm_cbam_output(llm_text):
    """LLM'den gelen CBAM maliyet tablosunu DataFrame'e √ßevirir"""
    cols = ['Quarter', 'ETS_Price', 'CBAM_Cost']

    if not isinstance(llm_text, str) or '|' not in llm_text:
        return pd.DataFrame(columns=cols)

    lines = [l.strip() for l in llm_text.split('\n') if '|' in l]
    parsed = []

    for line in lines:
        if 'quarter' in line.lower() or '---' in line:
            continue

        parts = [p.strip() for p in line.split('|')]
        if len(parts) < 3:
            continue

        try:
            parsed.append({
                'Quarter': parts[0],
                'ETS_Price': float(parts[1].replace('‚Ç¨','').replace(',','').replace('EUR','').strip()),
                'CBAM_Cost': float(parts[2].replace('‚Ç¨','').replace(',','').replace('EUR','').strip())
            })
        except:
            continue

    return pd.DataFrame(parsed, columns=cols)

# ------------------------------
# 2Ô∏è‚É£ Risk Analizi Ekle
# ------------------------------
def add_risk_analysis(cbam_df):
    """CBAM maliyetlerine g√∂re risk seviyeleri ekler"""
    if cbam_df.empty:
        cbam_df['Risk_Level'] = []
        return cbam_df

    q90 = cbam_df['CBAM_Cost'].quantile(0.9)
    mean_cost = cbam_df['CBAM_Cost'].mean()

    def risk_label(x):
        if x >= q90:
            return "üî¥ High Risk"
        elif x >= mean_cost:
            return "üü† Medium Risk"
        else:
            return "üü¢ Low Risk"

    cbam_df['Risk_Level'] = cbam_df['CBAM_Cost'].apply(risk_label)
    return cbam_df

# ------------------------------
# 3Ô∏è‚É£ Rapor ƒ∞√ßin Metrik Hesaplama
# ------------------------------
def calculate_metrics(cbam_summary, ets_forecast_table, cbam_df):
    """Rapor i√ßin t√ºm metrikleri hesaplar"""
    metrics = {}
    
    # √úr√ºn bilgileri
    metrics['product'] = cbam_summary['product']
    metrics['category'] = cbam_summary['category']
    metrics['quantity_tonnes'] = cbam_summary['quantity_tonnes']
    metrics['total_ei'] = cbam_summary['total_ei']
    metrics['current_cbam_cost'] = cbam_summary['cbam_cost']
    
    # ETS Forecast analizleri - DataFrame kontrol√º
    if isinstance(ets_forecast_table, pd.DataFrame) and not ets_forecast_table.empty:
        # 'Forecasted Value' veya 'Forecasted_Value' s√ºtununu bul
        ets_col = None
        if 'Forecasted Value' in ets_forecast_table.columns:
            ets_col = 'Forecasted Value'
        elif 'Forecasted_Value' in ets_forecast_table.columns:
            ets_col = 'Forecasted_Value'
        elif 'forecasted_value' in ets_forecast_table.columns:
            ets_col = 'forecasted_value'
        
        if ets_col:
            ets_prices = ets_forecast_table[ets_col]
            metrics['ets_min'] = ets_prices.min()
            metrics['ets_max'] = ets_prices.max()
            metrics['ets_avg'] = ets_prices.mean()
            metrics['ets_start'] = ets_prices.iloc[0]
            metrics['ets_end'] = ets_prices.iloc[-1]
            
            # Trend hesaplama
            change_pct = ((metrics['ets_end'] - metrics['ets_start']) / metrics['ets_start']) * 100
            if change_pct > 10:
                metrics['ets_trend'] = "‚ÜóÔ∏è Y√ºkseli≈ü"
            elif change_pct < -10:
                metrics['ets_trend'] = "‚ÜòÔ∏è D√º≈ü√º≈ü"
            else:
                metrics['ets_trend'] = "‚û°Ô∏è Stabil"
            metrics['ets_trend_pct'] = f"{change_pct:+.1f}%"
        else:
            metrics.update({
                'ets_min': 0, 'ets_max': 0, 'ets_avg': 0,
                'ets_start': 0, 'ets_end': 0,
                'ets_trend': "N/A", 'ets_trend_pct': "N/A"
            })
    else:
        metrics.update({
            'ets_min': 0, 'ets_max': 0, 'ets_avg': 0,
            'ets_start': 0, 'ets_end': 0,
            'ets_trend': "N/A", 'ets_trend_pct': "N/A"
        })
    
    # CBAM Cost analizleri
    if not cbam_df.empty:
        metrics['total_cbam_cost'] = cbam_df['CBAM_Cost'].sum()
        metrics['avg_cbam_cost'] = cbam_df['CBAM_Cost'].mean()
        metrics['max_cbam_cost'] = cbam_df['CBAM_Cost'].max()
        metrics['min_cbam_cost'] = cbam_df['CBAM_Cost'].min()
        
        # Kritik d√∂nem
        idx_max = cbam_df['CBAM_Cost'].idxmax()
        metrics['critical_quarter'] = cbam_df.loc[idx_max, 'Quarter']
        metrics['critical_cost'] = cbam_df.loc[idx_max, 'CBAM_Cost']
        
        # Yƒ±llƒ±k toplam
        cbam_df['Year'] = cbam_df['Quarter'].str.extract(r'(\d{4})')
        yearly = cbam_df.groupby('Year')['CBAM_Cost'].sum()
        metrics['yearly_totals'] = yearly.to_dict()
        metrics['highest_year'] = yearly.idxmax()
        metrics['highest_year_cost'] = yearly.max()
    else:
        metrics.update({
            'total_cbam_cost': 0, 'avg_cbam_cost': 0,
            'max_cbam_cost': 0, 'min_cbam_cost': 0,
            'critical_quarter': 'N/A', 'critical_cost': 0,
            'yearly_totals': {}, 'highest_year': 'N/A',
            'highest_year_cost': 0
        })
    
    return metrics

# ------------------------------
# 4Ô∏è‚É£ GEMINI ƒ∞LE RAPOR √úRETƒ∞Mƒ∞
# ------------------------------
def build_executive_report_prompt(cbam_summary, ets_forecast_table, cbam_df, metrics):
    """Gemini i√ßin y√∂netici raporu prompt'u olu≈üturur"""
    
    # ETS forecast tablosu
    if isinstance(ets_forecast_table, pd.DataFrame):
        ets_table_str = ets_forecast_table.to_string(index=False)
    else:
        ets_table_str = str(ets_forecast_table)
    
    # CBAM cost tablosu
    if not cbam_df.empty:
        cbam_table_str = cbam_df.to_string(index=False)
    else:
        cbam_table_str = "No data available"
    
    # Yearly breakdown
    yearly_text = "\n".join([f"{year}: ‚Ç¨{cost:,.0f}" for year, cost in metrics['yearly_totals'].items()])
    
    prompt = f"""
Sen bir **EU CBAM Finansal Danƒ±≈ümanƒ±**sƒ±n. A≈üaƒüƒ±daki verilere dayanarak **√ºst d√ºzey y√∂neticiler** i√ßin profesyonel bir rapor hazƒ±rla.

---
## üì¶ √úR√úN Bƒ∞LGƒ∞LERƒ∞
- **√úr√ºn**: {metrics['product']}
- **Sekt√∂r**: {metrics['category']}
- **ƒ∞thalat Miktarƒ±**: {metrics['quantity_tonnes']:,.0f} ton
- **Emisyon Yoƒüunluƒüu**: {metrics['total_ei']:.2f} tCO2/ton
- **Mevcut CBAM Maliyeti**: ‚Ç¨{metrics['current_cbam_cost']:,.2f}

---
## üìä ETS Fƒ∞YAT TAHMƒ∞NLERƒ∞ (Q1 2025 - Q4 2030)

{ets_table_str}

**ETS Fiyat ƒ∞statistikleri:**
- Minimum: ‚Ç¨{metrics['ets_min']:.2f}
- Maksimum: ‚Ç¨{metrics['ets_max']:.2f}
- Ortalama: ‚Ç¨{metrics['ets_avg']:.2f}
- Trend: {metrics['ets_trend']} ({metrics['ets_trend_pct']})

---
## üí∞ CBAM MALƒ∞YET PROJEKSƒ∞YONLARI

{cbam_table_str}

**CBAM Maliyet ƒ∞statistikleri:**
- **Toplam CBAM Maliyeti (2025-2030)**: ‚Ç¨{metrics['total_cbam_cost']:,.2f}
- **Ortalama √áeyreklik Maliyet**: ‚Ç¨{metrics['avg_cbam_cost']:,.2f}
- **En Y√ºksek Maliyet D√∂nemi**: {metrics['critical_quarter']} (‚Ç¨{metrics['critical_cost']:,.2f})

**Yƒ±llƒ±k CBAM Maliyet Daƒüƒ±lƒ±mƒ±:**
{yearly_text}

**En Y√ºksek Maliyetli Yƒ±l**: {metrics['highest_year']} (‚Ç¨{metrics['highest_year_cost']:,.2f})

---
## üéØ G√ñREV

A≈üaƒüƒ±daki ba≈ülƒ±klar altƒ±nda **y√∂netici raporu** hazƒ±rla:

### 1. EXECUTIVE SUMMARY (Y√∂netici √ñzeti)
- Toplam CBAM risk tutarƒ±
- Ana bulgular (2-3 c√ºmle)
- Kritik d√∂nemler

### 2. RISK ANALƒ∞Zƒ∞
- Y√ºksek riskli d√∂nemler
- ETS fiyat volatilitesi
- Maliyet artƒ±≈ü trendleri

### 3. STRATEJƒ∞K √ñNERƒ∞LER
- Kƒ±sa vadeli aksiyonlar (2025-2026)
- Orta vadeli aksiyonlar (2027-2028)
- Uzun vadeli aksiyonlar (2029-2030)

### 4. Fƒ∞NANSAL ETKƒ∞
- Yƒ±llƒ±k maliyet artƒ±≈üƒ±
- B√ºt√ße planlama √∂nerileri
- Nakit akƒ±≈üƒ± etkileri

### 5. SONU√á VE TAVSƒ∞YELER

---
**NOT**: Rapor T√ºrk√ße olmalƒ±, profesyonel ve net bir dille yazƒ±lmalƒ±. Rakamlarƒ± vurgula.
"""
    
    return prompt

# ------------------------------
# 5Ô∏è‚É£ ANA FONKSƒ∞YON - RAPOR √úRET
# ------------------------------
def generate_executive_cbam_report(cbam_summary, ets_forecast_table, cbam_cost_response, client):
    """
    Gemini ile profesyonel CBAM y√∂netici raporu olu≈üturur
    
    Args:
        cbam_summary: √úr√ºn √∂zet bilgileri (dict)
        ets_forecast_table: ETS fiyat tahmin tablosu (DataFrame veya string)
        cbam_cost_response: LLM CBAM maliyet √ßƒ±ktƒ±sƒ± (string)
        client: Gemini API client
    
    Returns:
        dict: metrics, cbam_df, report_text
    """
    
    print("\n" + "="*70)
    print("üìä CBAM Y√ñNETƒ∞Cƒ∞ RAPORU √úRETƒ∞Lƒ∞YOR...")
    print("="*70 + "\n")
    
    # 1. CBAM cost tablosunu parse et
    cbam_df = parse_llm_cbam_output(cbam_cost_response)
    
    # 2. Risk analizi ekle
    cbam_df = add_risk_analysis(cbam_df)
    
    # 3. Metrikleri hesapla
    metrics = calculate_metrics(cbam_summary, ets_forecast_table, cbam_df)
    
    # 4. Gemini prompt'u olu≈ütur
    report_prompt = build_executive_report_prompt(cbam_summary, ets_forecast_table, cbam_df, metrics)
    
    # 5. Gemini ile rapor √ºret
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=report_prompt
    )
    
    report_text = response.text
    
    # 6. Raporu yazdƒ±r
    print("\n" + "="*70)
    print(" CBAM Y√ñNETƒ∞Cƒ∞ RAPORU")
    print("="*70 + "\n")
    print(report_text)
    print("\n" + "="*70)
    print(" Rapor ba≈üarƒ±yla olu≈üturuldu")
    print("="*70 + "\n")
    
    return {
        'metrics': metrics,
        'cbam_df': cbam_df,
        'report_text': report_text
    }
```
